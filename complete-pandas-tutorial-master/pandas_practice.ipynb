ACCESSING DATA WITH PANDAS

```python
coffee.head()
```

```python
#Finding specfic values (loc & iloc)
#loc - rows, columns -> specfic with upper index
coffee.loc[0]
```

```python
coffee.loc[[0,1,2]]
```

```python
coffee.loc[5:8, ["Day","Units Sold"]]
```

```python
#iloc the same but uses indexes instead. no "Day" or "Units Sold" -> only uses index values [rows,columns]
coffee.iloc[0:5,[0,2]]
```

```python
coffee.head()
```

```python
#modifying a value
coffee.loc[1,"Units Sold"] = 10
```
